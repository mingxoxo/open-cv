```
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img = cv.imread('IU.jpg')

plt.imshow(img)
plt.xticks([]) # x축 눈금
plt.yticks([]) # y축 눈금
plt.show()
```

```
img
```

```
b, g, r = cv2.split(img)   # img파일을 b,g,r 로 분리
img2 = cv2.merge([r,g,b]) # b, r을 바꿔서 Merge
```

```
plt.imshow(img2)
plt.xticks([]) # x축 눈금
plt.yticks([]) # y축 눈금
plt.show()
```
